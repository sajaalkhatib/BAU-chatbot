```python
# ========================================================================
# ========================== Prerocess Text ==============================
# ========================================================================  

english_punctuations = string.punctuation
punctuations_list =english_punctuations
stop_words1 = list(set(stopwords.words('english')))  

def remove_stopwords1(art):
    art = ' '.join([word for word in art.split() if word not in stop_words1])
    return art
        
def remove_repeating(art):
     return re.sub(r'(.)\1+', r'\1', art)     # keep only 1 repeat

def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    text = remove_repeating(text)
    text = remove_emoji(text)
    text = remove_stopwords1(text)
    return text

# ========================================================================
# ======================== Tokanization Text =============================
# ======================================================================== 

def Tokanize_text(text):
    tokens = word_tokenize(text)
    text = ' '.join([word for word in tokens if word not in stop_words1])
    return text

# ========================================================================
# ========================= Normalize Text ===============================
# ========================================================================
def normalize_art_ar(art):
    art = art.split('\n')
    art = ' '.join(art)
    art = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?]))''','',art)
    art = re.sub(r"\xa0", "", art)   
    art = ''.join([char for char in art if not char.isdigit()])
    art = re.sub(r"u'\u064B'", "", art)
    art = re.sub(r"u'\u064C'", "", art)
    art = re.sub(r"u'\u064D'", "", art)
    art = re.sub(r"u'\u064E'", "", art)
    art = re.sub(r"u'\u064F'", "", art)
    art = re.sub(r"u'\u0650'", "", art)
    art = re.sub(r"u'\u08F0'", "", art)
    art = re.sub(r"u'\u08F1'", "", art)
    art = re.sub(r"u'\u08F2'", "", art)
    art = re.sub(r"_", " ", art)
    art = ''.join(c for c in art if not ud.category(c).startswith('P'))
    art = ''.join([char for char in art if not char in string.punctuation])
    return art

# ========================================================================
# ======================Load the Saving the model=========================
# ========================================================================

from keras.models import load_model
model = load_model('chatbot_model_v1.h5')
import json
import random
with open("dataintents.json", encoding="utf-16") as f:
     intents = json.load(f)
words = pickle.load(open('datawords.pkl','rb'))
classes = pickle.load(open('dataclasses.pkl','rb'))

# ========================================================================
# =================Clean the given Sentence(Qustion)======================
# ========================================================================

def clean_up_sentence(sentence):
    ######add text preprocessing#######
    xx=remove_stopwords1(sentence)
    xx=clean_text(xx)
    xx=normalize_art_ar(xx)
    ########end text preprocessing########
    sentence_words = nltk.word_tokenize(xx)#sentence)
    sentence_words = [lemmatizer.lemmatize(word.lower()) for word in sentence_words]
    return sentence_words

# return bag of words array: 0 or 1 for each word in the bag that exists in the sentence

# ========================================================================
# ===============Bag of words the given Sentence(Qustion)=================
# ========================================================================

def bow(sentence, words, show_details=True):
    # tokenize the pattern
    sentence_words = clean_up_sentence(sentence)
    # bag of words - matrix of N words, vocabulary matrix
    bag = [0]*len(words)
    for s in sentence_words:
        for i,w in enumerate(words):
            if w == s:
                # assign 1 if current word is in the vocabulary position
                bag[i] = 1
                if show_details:
                    print ("found in bag: %s" % w)
    return(np.array(bag))

# ========================================================================
# ============Predict Class of the given Sentence(Qustion)================
# ========================================================================

def predict_class(sentence, model):
    # filter out predictions below a threshold
    p = bow(sentence, words, show_details=False)
    res = model.predict(np.array([p]))[0]
    ERROR_THRESHOLD = 0.25
    results = [[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    # sort by strength of probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": classes[r[0]], "probability": str(r[1])})
    return return_list

# ========================================================================
# ===========Get the Answer of the given Sentence(Qustion)================
# ========================================================================

def getResponse(ints, intents_json):
    if not ints:
        return "I'm sorry, I couldn't understand your question."

    tag = ints[0]['intent']
    print(tag)
    list_of_intents = intents_json['intents']
    for i in list_of_intents:
        if i['tag'] == tag:
            result = random.choice(i['responses'])
            break
    else:
        # If no matching tag is found, provide a default response
        result = "I apologize, I do not have an answer to your question. See the department head, Dr. Atef Obeidat."

    return result

def chatbot_response(msg):
    ints = predict_class(msg, model)
    print(ints)
    if not ints:
        # Handle the case where no intents were predicted
        return "I apologize, I do not have an answer to your question. See the department head, Dr. Atef Obeidat."


    res = getResponse(ints, intents)
    return res
```
